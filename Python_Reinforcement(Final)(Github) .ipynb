Replacing the outliers using the IQR method

```python
numeric_col=house_data.select_dtypes(include=['number'])
for i in range(10):
    for col in numeric_col.columns:
        Q1 = house_data[col].quantile(0.25)
        Q3 = house_data[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5* IQR
        upper_bound = Q3 + 1.5* IQR
        
        print(f"Lower bound:{lower_bound}")
        print(f"Upper bound:{upper_bound}")
        #calculate the median
        median_value = house_data[col].median()
        print(f"median value:{median_value}")
        #replace outliers
        house_data.loc[(house_data[col]<=lower_bound) | (house_data[col]>=upper_bound),col]=round(median_value,0)
        print("after replacing outliers:",house_data)
```

Hypothesis Testing

     1. Descriptive Statistics

     3. two way T-Test: price vs is_renovated

Descriptive Stats

```python
# Summary statistics for key variables
desc_stats = house_data[['price', 'sqft_living', 'bedrooms', 'bathrooms', 'price_per_sqft']].describe()
print(desc_stats)
```

One way ANOVA test: price vs condition

```python
from scipy.stats import f_oneway

# Group price by condition
grouped_prices = [house_data[house_data['condition'] == c]['price'] for c in house_data['condition'].unique()]

# Perform one-way ANOVA
f_stat, p_value = f_oneway(*grouped_prices)

# Print results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis.")
    print("Conclusion: At least one property condition group has a significantly different average price.")
else:
    print("Fail to reject the null hypothesis.")
    print("Conclusion: No significant difference in average price across property conditions.")

```

The ANOVA test showed a significant difference in prices across property conditions.
Better condition homes tend to have higher average prices, affecting buyer decisions.

TWO WAY T-TEST

```python
from scipy.stats import ttest_ind

# Split data
renovated = house_data[house_data['is_renovated'] == 1]['price']
not_renovated = house_data[house_data['is_renovated'] == 0]['price']

# Perform t-test 
t_stat, p_val = ttest_ind(renovated, not_renovated, equal_var=False)

# Significance level
alpha = 0.05

# Print results
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_val:.5f}")

# Interpret results
if p_val < alpha:
    print(f"Since p-value ({p_val:.5f}) < alpha ({alpha}), we reject the null hypothesis.")
    print("Conclusion: There is a significant difference in mean price between renovated and not renovated homes.")
else:
    print(f"Since p-value ({p_val:.5f}) >= alpha ({alpha}), we fail to reject the null hypothesis.")
    print("Conclusion: No significant difference in mean price between renovated and not renovated homes.")
```

Chi-Square Test â€“ Renovation Status vs Property Condition

```python
from scipy.stats import chi2_contingency

# Create contingency table
contingency_table = pd.crosstab(house_data['is_renovated'], house_data['condition'])

# Perform Chi-Square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Output results
print("Chi-Square Statistic:", round(chi2, 2))
print("Degrees of Freedom:", dof)
print("P-Value:", round(p, 5))

if p < 0.05:
    print("Reject the null hypothesis - variables are not independent")
else:
    print("Fail to reject the null hypothesis - variables are independent")
```

Hâ‚€ (Null Hypothesis): Renovation status and condition rating are independent.

Hâ‚ (Alternative Hypothesis): Renovation status and condition rating are associated.


A low p-value (< 0.05) would suggest that renovated homes tend to have better condition ratings, which aligns with real-world expectations
Renovation status and condition are strongly associated
